```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('E:/Research_Files/JVF.csv')
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values
data.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
type_mapping = {'Not-J': 1, 'Cross': 2, 'T/Jun': 3, 'StagX': 4, 'Round': 5, 'RailW': 6, 'Other': 7}
```

```python
data['Junction Type'] = data['Junction Type'].map(type_mapping)
```

```python
X = data[['Junction Type']]
y = data['Fatal']
```

```python
model.fit(X, y)
```

```python
predicted_total_incidents = model.predict([[1]])
print("Predicted Fatal incidents for Junction Type Not-J:", predicted_total_incidents[0])
```

```python
plt.scatter(X, y, color='blue', label='Data Points')

plt.plot(X, model.predict(X), color='red', label='Linear Regression Line')
# Adding labels and title
plt.xlabel('Junction Type (Numerical)')
plt.ylabel('Total Number of Fatal Incidents')
plt.title('Scatter Plot: Fatal Incidents vs. Junction Type')

# Showing the plot
plt.legend()
plt.grid(True)
plt.show()
```
